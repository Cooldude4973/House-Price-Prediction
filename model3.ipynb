```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel(r'HousePricePrediction.xlsx')
```

```python
df.head()
```

```python
df_new =  df.drop(['MSSubClass' , 'MSZoning' , 'LotConfig' , 'Exterior1st' , 'BsmtFinSF2'] , axis=1)
```

```python
df_new.head()
```

```python
df_new = df_new.dropna(axis=0)
```

```python
df_new.head()
```

```python
df_new.shape
```

```python
df_new.info()
```

```python
df_new['BldgType'].unique()
```

```python
y = df_new.iloc[ : , -1]
df_new.drop(['SalePrice'] , axis=1 , inplace=True)
```

```python
X = df_new
```

```python
from sklearn.model_selection import train_test_split

train_X , test_X , train_Y , test_Y = train_test_split(X , y , test_size=0.2 , random_state= 0)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import OneHotEncoder

lr  = LinearRegression()
```

```python
ohe=OneHotEncoder()
ohe.fit(X[['BldgType']])
```

```python
from sklearn.compose import make_column_transformer

column_trans=make_column_transformer((OneHotEncoder(categories=ohe.categories_),['BldgType']),
                                    remainder='passthrough')
```

```python
from sklearn.pipeline import make_pipeline

pipe=make_pipeline(column_trans,lr)
```

```python
pipe.fit(train_X ,train_Y)
```

```python
y_pred=pipe.predict(test_X)
```

```python
from sklearn.metrics import r2_score

r2_score(test_Y,y_pred)
```

```python
scores=[]
for i in range(1000):
    X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.1,random_state=i)
    lr=LinearRegression()
    pipe=make_pipeline(column_trans,lr)
    pipe.fit(X_train,y_train)
    y_pred=pipe.predict(X_test)
    scores.append(r2_score(y_test,y_pred))
```

```python
np.argmax(scores)
```

```python
scores[np.argmax(scores)]
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.1,random_state=np.argmax(scores))
lr=LinearRegression()
pipe=make_pipeline(column_trans,lr)
pipe.fit(X_train,y_train)
y_pred=pipe.predict(X_test)
r2_score(y_test,y_pred)
```

```python
import pickle

pickle.dump(pipe , open('housePredictionLinearModel.pkl' , 'wb'))
```
